## Customer Data

```python
df = pd.read_csv(r"D:\Zephyr Data\project\OG Data\DimCustomer.csv",encoding="latin1")
```

```python
df
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
```

```python
# Split the cell to convert into float
df[["Currency", "Income"]] = df["YearlyIncome"].str.split(" ", expand = True)
```

```python
# Dropped unnecessary columns
df = df.drop(["CustomerAlternateKey","NumberCarsOwned","Phone","Title","NameStyle","Suffix","TotalChildren","EnglishEducation","SpanishEducation","FrenchEducation","EnglishOccupation","SpanishOccupation","FrenchOccupation","HouseOwnerFlag",'AddressLine2',"YearlyIncome"],axis=1)
```

```python
df.info()
```

```python
df['Income'] = df["Income"].astype("int")
```

```python
# Replaced unnecessary information from column
df["CommuteDistance"] = df["CommuteDistance"].str.replace(" Miles", "")
```

```python
df.duplicated().sum()
```

```python
df.to_csv(r"D:\Zephyr Data\project\Cleaned Data\Dim_Customer.csv", index= False)
```
